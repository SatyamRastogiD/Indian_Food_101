```python
words = {}
for i in df['ingredients']:
  for a in i.split(","):
    if a in words:
      words[a] += 1
    else:
      words[a] = 1
print(words)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
veggies = pd.get_dummies(new_df_2["diet"], drop_first = True)
```

```python
flavour = pd.get_dummies(new_df_2["flavor_profile"], drop_first = True)
```

```python
course = pd.get_dummies(new_df_2["course"], drop_first = True)
```

```python
new_df_3 = new_df_2.loc[:,new_df_2.columns.drop(["diet","flavor_profile","course"])]
```

```python
new_df_4 = pd.concat([new_df_3,veggies,flavour,course], axis = 1)
```

```python
new_df_4
```

```python
from sklearn.metrics.pairwise import cosine_similarity
new_df_4.set_index('name', inplace = True)
```

```python
new_df_4.iloc[:,1:]
```

```python
cos_sim = cosine_similarity(new_df_4.iloc[:,1:], new_df_4.iloc[:,1:])
```

```python
cos_sim.shape
```

```python
cos_sim_df = pd.DataFrame(cos_sim, index = new_df_4.index, columns= new_df_4.index.values)
```

```python
cos_sim_df
```

```python
def similar_dishes(i):
  if i in cos_sim_df.index.values:
    return(cos_sim_df[i].sort_values(ascending = False).head(5))
  else:
    return("No such dish in the data")
```
